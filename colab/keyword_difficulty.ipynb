```python
import pandas as pd
import ast
import random
df = pd.read_csv('../drive/Shareddrives/CNPM_mr/temp')
l = []
for c in df.title:
  c = ast.literal_eval(c)
  l += c
keywords = random.sample(l,100)
print(keywords)
```

```python
from client import RestClient

client = RestClient("vinh.huynhdavid2002@hcmut.edu.vn", "77251e7a9690a9a2")
post_data = dict()
# simple way to set a task
post_data[len(post_data)] = dict(
    keywords=keywords,
    location_name="Vietnam",
    language_name="Vietnamese"
)
# POST /v3/dataforseo_labs/google/bulk_keyword_difficulty/live
response = client.post("/v3/dataforseo_labs/google/bulk_keyword_difficulty/live", post_data)
# you can find the full list of the response codes here https://docs.dataforseo.com/v3/appendix/errors
if response["status_code"] == 20000:
    print(response)
    # do something with result
else:
    print("error. Code: %d Message: %s" % (response["status_code"], response["status_message"]))
```

```python
result = response['tasks'][0]['result'][0]['items']
for c in result:
  print(c)
```
